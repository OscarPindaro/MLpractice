# Bayesian Inference and Conjugate priors

## Batch processing with bayes theorem

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

Creation of data extracted from a binomial with $p = 0.3$. In reality, these data are not generated but obtained from the field.

```python
success_prob = 0.3
data = np.random.binomial(n=1, p=success_prob, size = 1000)
```

We are interested in find the real succes probability of our Bernulli process. We need to make an assumption on the initial success probability. A safe assumption is to consider the success probability (here marked as $\theta$) as an uniform between $0$ and $1$. In this way, no bias will affect the process.

```python
theta_range = np.linspace(0,1, 1000)
theta_range_e = theta_range + 0.001
```

**Prior** is modeled with a *Beta* distribtution with $\alpha = 1$ and $\beta = 1$.

```python
a = 1
b = 1
prior = stats.beta.pdf(x = theta_range, a = a, b = b)
priorProb = stats.beta.cdf(x= theta_range_e, a = a, b = b) -stats.beta.cdf(x= theta_range, a = a, b = b)
```

```python
plt.rcParams['figure.figsize']= [13,4]
fig, ax = plt.subplots()
plt.plot(theta_range, prior, linewidth=4, color="gold")
plt.xlabel('theta', fontsize=16)
plt.ylabel('Density', fontsize=16)
plt.grid(alpha=0.4, linestyle='--')

plt.title("[Prior] PDF of initial prior")
plt.show()

fig, ax = plt.subplots()
plt.xlabel('theta', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.grid(alpha=0.4, linestyle='--')
plt.title("[Prior] PMF of initial prior")
plt.plot(theta_range, priorProb, linewidth=4, color="red")
plt.show()
```

Now we need to model the **likelyhood**. Since out data is a sequence of 0 and 1, we need to model it by using a binomial distribution (**DISCRETE!!!!!**). For only 1 k, the pmf returns the probability of that combination of successes and failures. If k is an np.array, the likelihood will be a probability wrt the number of successes. Here we are calculating the likelihood with p = 0.7 that we have, but can be multiplied by any value of p. If p slide from 0 to 1, the peak of the distribution will move as well, and the point in th epeak is the number of successes most probable with said p.

```python
K = np.arange(0, len(data)) #generate all possible k, number of success
likelyhood = stats.binom.pmf(n=len(data), k=K, p=0.7) 
```

```python
plt.plot(K, likelyhood, 'b')
plt.title("[Likelihood] Probability of observing k successes with {} success probability".format(0.7), fontsize=22)
plt.xlabel("K", fontsize = 16)
plt.ylabel("Probability", fontsize = 16)
plt.grid(alpha=.4, linestyle='--')
plt.show()
```

We need now  to calculate the **Posterior**. We will use the *Bayes Theorem*. the following line is the probability of observing this particular data given a ceratin success probability. I'm doing this for EVERY success probability (that's why theta range is assigned to the probability). In this way, for p=0.3, i check the probability of oserving data give p = 0.3, and the same for 0.35, 0.36, 0.37 (i obtain an array that has a probability of observing data for wach possible success rate).<br>
$P(\theta | n = n_{0}, k = k_{0} ) = \frac{P(n =n_{0}, k = k_{0} | \theta) * P(\theta)} {\alpha}$, where $\alpha$ is a marginalization constant. <br>
$\alpha$ is in particular the integral of the numerator over $\theta$. In fact, i'm interested to see how much the *probability area* of the numerator is in proportion of the probability of observing X. (rephrasing necessario)

```python
likelyhood = stats.binom.pmf(n=len(data), k=np.sum(data), p=theta_range) 
```

```python
plt.plot(theta_range, likelyhood, 'b')
plt.title("[Likelihood] Probability of observing k = {} successes wrt success probability".format( np.sum(data)), fontsize=22)
plt.xlabel("K", fontsize = 16)
plt.ylabel("Probability", fontsize = 16)
plt.grid(alpha=.4, linestyle='--')
plt.show()
```

The following cells rewrites all the quantities needed, so that it's clearer.

```python
a=1
b = 1
priorProb = stats.beta.cdf(x= theta_range_e, a = a, b = b) -stats.beta.cdf(x= theta_range, a = a, b = b)
priorProb[:-1] = priorProb[10] #this way the plot is better
likelyhood = stats.binom.pmf(n=len(data), k=np.sum(data), p=theta_range) 
posterior = likelyhood * priorProb
posteriorN = posterior / posterior.sum()
```

Now i will plot the *prior*, the *likelihood* and the *posterior*, in order to appreciate the shift of the beta function towards the data.

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(20,7))
plt.xlabel('p', fontsize=24)

#prior
axes[0].plot(theta_range, priorProb, label="Prior", linewidth = 3, color ="red")
axes[0].set_title("Prior", fontsize=16)
#likelihood
axes[1].plot(theta_range, likelyhood, label="Likeliood", linewidth = 3, color ="yellow")
axes[1].set_title("Likelihood", fontsize=16)
#posterior
axes[2].plot(theta_range, posteriorN, label="Posterior", linewidth = 3, color ="blue")
axes[2].set_title("Posterior", fontsize=16)

plt.show()
```

Now suppose that another batch of data arrives. Let's suppose also that this time the data comes from a very lucky distribution, with hai success rate.

```python
priorProb = posteriorN
likelyhood = stats.binom.pmf(n = 5000, k = 4000, p = theta_range)
posterior = likelyhood*priorProb
posteriorN = posterior / posterior.sum()
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(20,10))
plt.xlabel('p', fontsize=24)

#prior
axes[0].plot(theta_range, priorProb, label="Prior", linewidth = 3, color ="red")
axes[0].set_title("Prior", fontsize=16)
#likelihood
axes[1].plot(theta_range, likelyhood, label="Likeliood", linewidth = 3, color ="gold")
axes[1].set_title("Likelihood", fontsize=16)
#posterior
axes[2].plot(theta_range, posteriorN, label="Posterior", linewidth = 3, color ="blue")
axes[2].set_title("Posterior", fontsize=16)

plt.show()
```

The estimation of the success rate is the one with the higest probability (MAP).

```python
result = np.argmax(posteriorN)
print("The success rate of this bernulli process is {}".format(result/1000))
```

## Conjugate priors

The drawback of using the previous approach was that each time a batch of data comes in, a product and a marginalization have to be done, degrading performance. In this particular case, since we are using binomals and Beta distribution, a close form solution can be used.

```python
a = 2
b = 8
dp= 100000
p_range = np.linspace(0,1,dp)
prior = stats.beta.pdf(p_range, a=a, b=b)
```

Let's suppose that now we estract 2000 data, and we obtain 1300 successes. The posterior will just update the values of a and b in the Beta distribution.

```python
N = 2000
k = 1300
newA = a + 1300
newB = b + N -k
posterior = stats.beta.pdf(p_range, a=newA, b = newB)
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(20,7))
plt.xlabel('p', fontsize=24)

#prior
axes[0].plot(p_range, prior, label="Prior".format(a,b), linewidth = 3, color ="red")
axes[0].set_title("Prior with a={} , b={}".format(a,b), fontsize=16)
#posterior
axes[1].plot(p_range, posterior, label="Posterior", linewidth = 3, color ="blue")
axes[1].set_title("Posterior with a={} and b={}".format(newA, newB),fontsize=16)

plt.show()
```

## Application with wierd probabilities

```python
import math
prior = np.zeros(1000)
prior[0:200] = np.linspace(3, 10000, 200)
prior[200: len(prior)] = np.linspace(100, 2 ,800)
prior = prior/prior.sum()
p_range = np.linspace(0,1,1000)
```

```python
plt.rcParams['figure.figsize']= [7,2]
fig, ax = plt.subplots()
plt.plot(p_range, prior, linewidth=2, color="red")
plt.xlabel('p', fontsize=16)
plt.ylabel('Density', fontsize=12)
plt.grid(alpha=0.4, linestyle='--')

plt.title("[Prior] PMF of initial prior")
plt.show()
```

```python
likelyhood = stats.binom.pmf(k = 10, n=30, p=p_range)
posterior = likelyhood*prior
posterior = posterior/posterior.sum()
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(7,7))
plt.xlabel('p', fontsize=24)

#prior
axes[0].plot(p_range, prior, label="Prior", linewidth = 3, color ="red")
axes[0].set_title("Prior", fontsize=16)
#likelihood
axes[1].plot(p_range, likelyhood, label="Likelyhood", linewidth = 3, color ="gold")
axes[1].set_title("Likelyhood", fontsize=16)
#posterior
axes[2].plot(p_range, posterior, label="Posterior", linewidth = 3, color ="blue")
axes[2].set_title("Posterior", fontsize=16)

plt.show()
```
